For this project __Multi Agent with DDPG__ was implemented with following hyperparameter
    - State size = 48
    - BUFFER_SIZE = int(5e5) (replay buffer size)
    - BATCH_SIZE = 128       (minibatch size)
    - GAMMA = 0.99           (discount factor)
    - TAU = 5e-3             (for soft update of target parameters)
    - LR_ACTOR = 1e-3        (learning rate of the actor)
    - LR_CRITIC = 1e-3       (learning rate of the critic)
    - WEIGHT_DECAY = 0       (L2 weight decay)
    
    The 2 Models Actor and Critic uses the following architecture
     - Actor
       It uses 2 layers with 256 and 128 units and a Relu activation function and a tanh activation function for the output layer.
      - Critic
       It uses 2 layers with 256 and 256+ action state size to 128 units to the output layer, and a Relu activation function and no activation function in the output layer.
       
   Also, the __OUNice function__ was implemented for maximizing exploration and overall robust and quick learning.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

episode_scores = np.load('./episode_scores.npy')
average_scores = np.load('./average_scores.npy')
plt.figure(figsize=(12,8))
plt.plot(episode_scores, label = 'Episode scores',)
plt.plot(average_scores, label = 'Average scores',)
plt.axhline(y=.5, color='r', linestyle='dashed')
plt.xlabel('Episodes')
plt.ylabel('Scores')
plt.title('Score Plot')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

The Agent reached the goal score of +.50 within 1600-1650 episodes. Although the average score was fluctuating it remained in a state of growth indicating that the model was learning and improving.
